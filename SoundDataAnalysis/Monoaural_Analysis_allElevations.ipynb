```python
# This code analyzes the Power Spectral Density for sounds
# from different directions (0 - 360 azimuth angle)

import os
import matplotlib.pyplot as plt
import numpy as np
import librosa
import wave
import librosa as lr
from scipy import signal
from scipy.fftpack import fft,fft2, fftshift, ifft
import math
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from sklearn import preprocessing
import librosa.display
import pandas as pd
```

```python
T=10
D=5
N=10
shift = 1/4   # number of cycles to shift (1/4 cycle)
x = np.linspace(0, T*N, 10000, endpoint=False)
y=signal.square(2 * np.pi * (1/T) * x + 2*shift*np.pi)
plt.plot(x,y)
plt.ylim(-2, 2)
plt.xlim(0, T*N)
```

```python
# For 50  degrees Elevation angle

directorySound_content = os.listdir("/home/prabhjot/Desktop/SoundData/OaklandU_Bridge")
directorySound = os.fsencode("/home/prabhjot/Desktop/SoundData/OaklandU_Bridge")

index = directorySound_content.index('1Meter_Azimuth210_Elevations')
_1MeterFolderPath_Az210 = (os.path.join(directorySound.decode("utf-8"),directorySound_content[index] ))
_1MeterFolderPath_Az210_content = os.listdir(_1MeterFolderPath_Az210)
subdir = (os.path.join(_1MeterFolderPath_Az210,_1MeterFolderPath_Az210_content[5]))
print(subdir)
    
    #Read the audio files
ch1, sfreq = lr.load(os.path.join(subdir, 'ch1.wav'), sr=44100)
ch2, sfreq = lr.load(os.path.join(subdir, 'ch2.wav'), sr=44100)
ch3, sfreq = lr.load(os.path.join(subdir, 'ch3.wav'), sr=44100)
RawAudio, sfreq =  lr.load(os.path.join(subdir, 'AUGUST_1_raw.wav'), sr=44100)

## Fix the length of the input to 2 seconds
RawAudio_fixedLength = librosa.util.fix_length(RawAudio, 2*sfreq)
# plt.figure(figsize=(14, 5))
# librosa.display.waveplot(RawAudio_fixedLength, sr=sfreq)
# plt.show()

## Convert to frequency domain by taking the Fourier transfrom
signal_fft = np.fft.fft(RawAudio_fixedLength)
magnitude = np.abs(signal_fft)
magnitude = 20*np.log10(magnitude)
frequency = np.linspace(0, sfreq, len(magnitude))

M = 50
#calculate moving average
Moving_Average = pd.Series(magnitude).rolling(window=M).mean().iloc[M-1:].values
print(Moving_Average)

# plt.figure(figsize=(20,7))
# plt.plot(frequency[0:], magnitude[0:]) 
# plt.plot(frequency[M-1:],Moving_Average[0:])
# plt.xlabel("Frequency (Hz)")
# plt.ylabel("Magnitude")
# plt.title("Magnitude spectrum")
# plt.show()


index_4K = (np.where(frequency>=4000))
Num_samples = 3000
plt.figure(figsize=(20,7))
plt.plot(frequency[index_4K[0][0]: index_4K[0][0]+Num_samples], magnitude[index_4K[0][0]: index_4K[0][0]+Num_samples]) 
plt.plot(frequency[index_4K[0][0]: index_4K[0][0]+Num_samples],Moving_Average[index_4K[0][0]: index_4K[0][0]+Num_samples])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude in dB")
plt.title("Magnitude spectrum")
plt.show()
```
